```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12, 8)
plt.rcParams.update({'font.size': 14})
```

```python
def gamblers_ruin():
    gambling_money = 50;
    gambling_goal = 100;
    gambling_simulations = [];

    while gambling_money in range(1, gambling_goal):
        bet_size = 1
        w_or_l = random.randrange(-1,2,step=2)
        gambling_money += bet_size * w_or_l
        gambling_simulations.append(gambling_money)
    return gambling_simulations
```

```python
plt.plot(gamblers_ruin())
```

```python
def prob_of_ruin(gambling_goal, initial_gambling_money):
    return (gambling_goal - initial_gambling_money)/ gambling_goal
```

```python
prob_of_ruin(100,90)
```

```python
mc_example = { "pizza": [0.25, 0, 0.75,1],
                "burger": [0.25, 0.25, 0, 0],
                "bun": [0.25,0.25,0.25,0],
                "cake": [0.25,0.5, 0, 0]}
mc = pd.DataFrame(data= mc_example, index=["pizza", "burger", "bun", "cake"])
```

```python
travel_sim = []
travel_sim.append(mc.iloc[0].index[0])
food = np.random.choice(mc.iloc[0].index, p=mc.iloc[0])
travel_sim.append(food)
```

```python
while len(travel_sim) <25:
    food = np.random.choice(mc.iloc[mc.index.get_loc(food)].index,p=mc.iloc[mc.index.get_loc(food)])
    travel_sim.append(food)
```

```python
def matrix_power(matrix, power):
    if power == 0:
        return np.identity(len(matrix))
    elif power == 1:
        return matrix
    else:
        return np.dot(matrix, matrix_power(matrix, power-1))
```
